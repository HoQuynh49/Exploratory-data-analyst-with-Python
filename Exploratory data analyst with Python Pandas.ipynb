# Step 2: Data Preparation

```python
# Dropping irrelevant columns and rows
# Identifying duplicated columns
# Renaming Columns
# Feature Creation
```

```python
df_new = df[['coaster_name', 
   #'Length', 'Speed', 
       'Location', 'Status', 
   # 'Opening date','Type', 
         'Manufacturer', 
   #'Height restriction', 'Model', 'Height',
       # 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
       # 'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
       # 'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
       # 'Track layout', 'Fastrack available', 'Soft opening date.1',
       #'Closing date', 'Opened', 'Replaced by', 'Website',
       # 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
       # 'Single rider line available', 'Restraint Style',
       # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced', 'latitude', 'longitude', 'Type_Main',
       'opening_date_clean', 
       # 'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', 
       # 'height_value', 'height_unit', 
       'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df_new['opening_date_clean'] = pd.to_datetime(df_new['opening_date_clean'])
df_new['opening_date_clean']
```

```python
df_new = df_new.rename(columns={
    'coaster_name': 'Coaster_name',
    'year_introduced' : 'Year_introduced',
    'latitude' : 'Latitude',
    'longitude' : 'Longitude',
    'Type_Main' : 'Type_main',
    'opening_date_clean' : 'Opening_date',
    'speed_mph' : 'Speed_mph',
    'height_ft' : 'Height_ft',
    'Inversions_clean' : 'Inversions',
    'Gforce_clean' : 'Gforce'
    
})
df_new.head(5)
```

```python
df_new.isna().sum()
```

```python
df_new.loc[df_new.duplicated()]
```

```python
df_new.loc[df_new.duplicated(subset=['Coaster_name'])]
```

```python
# checking an example of duplicate
df_new.query('Coaster_name == "Crystal Beach Cyclone"')
# we have some faults in Year_introduced' columns 
# -> so we will check the data by 'Coaster_name', 'Location', 'Opening_date_clean' to find the duplicates
```

```python
df_new = df_new.loc[~df_new.duplicated(subset = ['Coaster_name', 'Location', 'Opening_date_clean'])].reset_index(drop=True)
df_new.shape
```

# STEP 3: Feature Understanding

```python
# Plotting Feature Distributions
# Histogram
# KDE
# Boxplot
```

```python
year_intro = df_new['Year_introduced'].value_counts().head(10) \
            .plot(kind = 'bar', title = 'Top 10 Years Coaster Introduced')
year_intro.set_xlabel('Year Introduced')
year_intro.set_ylabel('Count')
```

The largest number of roller coasters introduced is in 1999.

```python
Speed = df_new['Speed_mph'].plot(kind = 'hist', bins = 20, title = 'Coaster Speed (mph)')
Speed.set_xlabel('Speed (mph)')
```

The majority of roller coasters have the speed from 50mph to 60 mph.

```python
Speed = df_new['Speed_mph'].plot(kind = 'kde', title = 'Coaster Speed (mph)')
Speed.set_xlabel('Speed (mph)')
```

# STEP 4: Feature Relationships

```python
# Scatterplot
# Heatmap Correlation
# Pairplot
# Groupby comparisons
```

```python
speed_vs_height = df_new.plot(kind = 'scatter', x='Speed_mph', y='Height_ft', title = 'Coaster Speed vs. Height')
plt.show()
```

As we can see, we have a trend that the higher the roller coaster is, the faster the speed is.

```python
speed_vs_height = sns.scatterplot(x='Speed_mph', y='Height_ft', hue='Year_introduced', data=df_new)
speed_vs_height.set_title('Coaster Speed vs. Height')
plt.show()
```

```python
sns.pairplot(df_new,
             vars=['Year_introduced','Speed_mph',
                   'Height_ft','Inversions','Gforce'],
            hue='Type_main')
plt.show()
```

```python
df_new_corr = df_new[['Year_introduced','Speed_mph',
                   'Height_ft','Inversions','Gforce']].dropna().corr()
df_new_corr
```

```python
sns.heatmap(df_new_corr, annot=True)
```

```python
# What are the location with fastest roller coaster (minimum of 10)?
ax = df_new.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean','count']) \
    .query('count >= 10') \
    .sort_values('mean')['mean'] \
    .plot(kind='barh', figsize=(12, 5), title='Average Coast Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
